# PyCitySchools Analysis

## Introduction

Could the PyCitySchools education system be using its resources more effectively? To answer this question, I analyzed two sets of data:  
a csv containing information on all of the students in the system,  
and a csv containing information on all of the schools in the system.

## Data Cleanup

Cleanliness was not an issue with the data sets.  
Both data sets began with 0 missing values, and all data types were correct from the beginning.
I did omit some of the original data provided:  
1) the 'gender' of the student.
2) the 'Student ID' of the student.
3) the 'School ID' of the school.

No further omissions or changes were made to the data in order to clean it.

## Analysis

The analysis for this data was performed in the following order:
1. The two data sets were merged together on the 'school name'.
2. A few calculations were made to create more data points on the district as a whole as follows:
    - the total amount of students in the district.
    - the amount of schools in the district.
    - the total budget for the district.
    - the average math score of students in the district and the percentage of students passing math.
    - the average reading score of students in the district and the percentage of students passing reading.
    - the overall passing rate of students in the district (students with a score of 70% or higher in both subjects listed above)
3. These calculations were used in a snapshot of the district as a whole.
4. A similar table to the district snapshot was created with the same data points for each individual school.
5. The passing rates of students for both subjects was broken into a grade-by-grade analysis.
6. A comparison of student passing rates and budget per capita was created.
7. A comparison of student passing rates and school size was created (with schools split into small, medium, or large).
8. A comparison of student passing rates and the type of school (district or charter) was created.
9. A leaderboard of the highest-performing schools was created.
10. A leaderboard of the lowest-performing schools was created.

## Observations

1. It appears that there is a negative correlation between the amount of the budget spent on students and the success rates of those students.
    - In the section [Scores by Student Spending](#scores-spending) on budget used per capita vs. success rate of students: the largest spending group returns with the lowest success rate of students. Looking further into this issue, we can see the nuance that is present: larger schools, which have the lowest success rates of students as noted in [Scores by School Size](#scores-size), are also the schools which are able to spend the most on their students. However: it appears this money is not being used effectively.
    - The data presented above suggests that the highest area of concern among large schools is mathematics. Large schools have a 24% lower passing rate in the subject (whereas the reading scores are behind 10% from the other schools); with that in mind, it could be beneficial to the students of larger schools if resources were allocated towards programs and assistance in mathematics. Given that the average score in math is 69.96% for the large schools: a mere 1% increase would improve the overall passing rates for these schools greatly.
2. Charter schools see higher success rates for students across the board.
   - In the section [Highest-Performing Schools](#highest-scores), the top 8 of 15 schools are charter schools. These charter schools are among the lowest in terms of budget available per student; yet, as in the section [Scores by School Type](#scores-type), the charter schools have a 30% higher overall rate of passing than the district schools. Given this information, I believe it would make sense to create programs for the district schools that are similar to the programs offered in the charter schools because it appears their effectiveness is high. 

## Conclusion

If the district schools are able to mimic programs offered in charter schools there could be a great increase in the rate of passing for students of district schools.

An area of opportunity exists within the mathematics department: increasing scores there (specifically within the larger schools) would lead to greatly increased passing rates across the board in the district.

# PyCitySchools Data

## Setup

```python
# Dependencies.
import pandas as pd
from pathlib import Path

# File paths for our CSV files.
school_path = Path("Resources/schools_complete.csv")
student_path = Path("Resources/students_complete.csv")

# Create DataFrames from our CSV files.
school_data = pd.read_csv(school_path)
student_data = pd.read_csv(student_path)

# Combine DataFrames into one.
school_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

# Is it working?
school_df.head()
```

```python
# Check out the Data for null/missing values and data types.
# As well as... just look at it a little before diving in!
display(school_df.shape)
display(school_df.info())
```

## District Summary

```python
# Calculate the total number of unique schools.
school_count = school_df["school_name"].nunique()
school_count
```

```python
# Calculate the total number of students.
student_count = school_df["Student ID"].nunique()
student_count
```

```python
# Calculate the total budget.
total_budget = sum(school_df["budget"].unique())
total_budget
```

```python
# Calculate the average math score.
avg_math_score = school_df["math_score"].mean()
avg_math_score
```

```python
# Calculate the average reading score.
avg_read_score = school_df["reading_score"].mean()
avg_read_score
```

```python
# Calculate the percentage of students who passed math.
passing_math_count = school_df[(school_df["math_score"] >= 70)].count()["Student ID"]
passing_math_perc = passing_math_count / float(student_count)
passing_math_perc
```

```python
# Calculate the percentage of students who passed reading.
passing_read_count = school_df[(school_df["reading_score"] >= 70)].count()["Student ID"]
passing_read_perc = passing_read_count / float(student_count)
passing_read_perc
```

```python
# Calculate the percentage of students that passed both math and reading.
overall_passing_count = school_df[(school_df["math_score"] >= 70) & 
                                  (school_df["reading_score"] >= 70)].count()["Student ID"]

overall_passing_rate = overall_passing_count / float(student_count)
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame.
district_summary = pd.DataFrame([{"Total Schools": school_count,
                                  "Total Students": student_count,
                                  "Total Budget": total_budget,
                                  "Average Math Score": avg_math_score,
                                  "Average Reading Score": avg_read_score,
                                  "% Passing Math": passing_math_perc,
                                  "% Passing Reading": passing_read_perc,
                                  "% Overall Passing": overall_passing_rate}])

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:,.2%}".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:,.2%}".format)
district_summary["% Overall Passing"] = district_summary["% Overall Passing"].map("{:,.2%}".format)

# Display the DataFrame.
district_summary
```

## School Summary

```python
# Select the type per school from school_data.
school_types = school_data.set_index(["school_name"])["type"]
school_types
```

```python
# Calculate the total student count per school from school_data.
per_school_counts = school_data.set_index(["school_name"])["size"]
per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school from school_data.
per_school_budget = school_data.set_index(["school_name"])["budget"]
per_school_capita = per_school_budget / per_school_counts
per_school_capita
```

```python
# Calculate the average test scores per school from school_df.
per_school_read = school_df.groupby(["school_name"])["reading_score"].mean()
per_school_math = school_df.groupby(["school_name"])["math_score"].mean()

display(per_school_read)
display(per_school_math)
```

```python
# Calculate the number of students per school with math scores of 70 or higher from school_df.
stu_passing_math = school_df[(school_df["math_score"] >= 70)].groupby(["school_name"]).size()
stu_passing_math
```

```python
# Calculate the number of students per school with reading scores of 70 or higher from school_df.
stu_passing_read = school_df[(school_df["reading_score"] >= 70)].groupby(["school_name"]).size()
stu_passing_read
```

```python
# Calculate the number of students per school passing both reading and math.
stu_passing_both = school_df[(school_df["reading_score"] >= 70) & 
                             (school_df["math_score"] >= 70)].groupby(["school_name"]).size()
stu_passing_both
```

```python
# Calculate the passing rates.
per_school_pass_math = stu_passing_math / per_school_counts
per_school_pass_read = stu_passing_read / per_school_counts
per_school_pass_both = stu_passing_both / per_school_counts
```

<a id="school-summary"><a>

```python
# Create a DataFrame called 'per_school_summary' with columns for the calculations above.
per_school_summary = pd.DataFrame({"School Type": school_types,
                                   "Total Students": per_school_counts,
                                   "Total School Budget": per_school_budget,
                                   "Per Student Budget": per_school_capita,
                                   "Average Math Score": per_school_math,
                                   "Average Reading Score": per_school_read,
                                   "% Passing Math": per_school_pass_math,
                                   "% Passing Reading": per_school_pass_read,
                                   "% Overall Passing": per_school_pass_both})
per_school_summary.index.name = None

# Make a 'pretty' version to display so that we can still modify the 'raw' version later for numbers.
pretty_school_summary = per_school_summary.copy()

# Formatting.
pretty_school_summary["Total School Budget"] = pretty_school_summary["Total School Budget"].map("${:,.2f}".format)
pretty_school_summary["Per Student Budget"] = pretty_school_summary["Per Student Budget"].map("${:,.2f}".format)
pretty_school_summary["% Passing Math"] = pretty_school_summary["% Passing Math"].map("{:,.2%}".format)
pretty_school_summary["% Passing Reading"] = pretty_school_summary["% Passing Reading"].map("{:,.2%}".format)
pretty_school_summary["% Overall Passing"] = pretty_school_summary["% Overall Passing"].map("{:,.2%}".format)

# Display the 'pretty' DataFrame.
pretty_school_summary
```

## Scores by School Spending
<a id="scores-spending"><a>

```python
# Establish the bins.
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget".
school_spending_df = per_school_summary.copy()
```

```python
# Use 'pad.cut' to categorize spending based on the bins.
school_spending_df['Spending Ranges (Per Student)'] = pd.cut(school_spending_df['Per Student Budget'], 
                                                            bins=spending_bins, 
                                                            labels=labels, 
                                                            include_lowest=True)
school_spending_df
```

```python
# Calculate averages for the desired columns.
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame.
spending_summary = pd.DataFrame({"Average Math Score": spending_math_scores, 
                                 "Average Reading Score": spending_reading_scores, 
                                 "% Passing Math": spending_passing_math, 
                                 "% Passing Reading": spending_passing_reading, 
                                 "% Overall Passing": overall_passing_spending})
# Format results.
spending_summary["Average Math Score"] = spending_summary["Average Math Score"].map("{:,.2f}".format)
spending_summary["Average Reading Score"] = spending_summary["Average Reading Score"].map("{:,.2f}".format)
spending_summary["% Passing Math"] = spending_summary["% Passing Math"].map("{:,.2%}".format)
spending_summary["% Passing Reading"] = spending_summary["% Passing Reading"].map("{:,.2%}".format)
spending_summary["% Overall Passing"] = spending_summary["% Overall Passing"].map("{:,.2%}".format)

# Display results.
spending_summary
```

## Scores by School Size
<a id="scores-size"><a>

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins.
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.
per_school_summary["School Size"] = pd.cut(per_school_summary['Total Students'], 
                                           bins = size_bins, 
                                           labels=labels, 
                                           include_lowest=True)
per_school_summary
```

```python
# Calculate averages for the desired columns.
size_math_scores = per_school_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Overall Passing"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`.
size_summary = pd.DataFrame({"Average Math Score": size_math_scores, 
                             "Average Reading Score": size_reading_scores, 
                             "% Passing Math": size_passing_math, 
                             "% Passing Reading": size_passing_reading, 
                             "% Overall Passing": size_overall_passing})

# Format results.
size_summary["Average Math Score"] = size_summary["Average Math Score"].map("{:,.2f}".format)
size_summary["Average Reading Score"] = size_summary["Average Reading Score"].map("{:,.2f}".format)
size_summary["% Passing Math"] = size_summary["% Passing Math"].map("{:,.2%}".format)
size_summary["% Passing Reading"] = size_summary["% Passing Reading"].map("{:,.2%}".format)
size_summary["% Overall Passing"] = size_summary["% Overall Passing"].map("{:,.2%}".format)

# Display results.
size_summary
```
